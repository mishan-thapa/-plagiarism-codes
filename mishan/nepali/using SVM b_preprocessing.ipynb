# dummy dataset

```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt

# create a toy dataset
X = np.array([-2, -1, 0, 1, 2]).reshape(-1, 1) # one feature
Y = np.array([0, 0, 1, 1, 1]) # binary labels

# fit a linear SVM model
clf = svm.SVC(kernel='linear', C=1)
clf.fit(X, Y)

# get the hyperplane parameters
w = clf.coef_[0]
b = clf.intercept_[0]
slope = -w[0]# / w[1]
intercept = -b / w[0]

# plot the dataset and hyperplane
plt.scatter(X[:, 0], y, color='b')
plt.plot(X[:, 0], clf.predict(X), color='r')
plt.plot(X[:, 0], slope*X[:, 0] + intercept, color='g')
plt.xlabel('Feature')
plt.ylabel('Label')
# set the y-axis limit to 0 and 1
plt.ylim((-0.1, 1.1))
plt.show()
```
