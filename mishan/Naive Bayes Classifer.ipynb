```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('final.csv')

```

```python
df.head(600)
```

```python
X = df[['trigram_similarity','word_similarity','bigram_similarity','lsa_similarity','fingerprint_similarity']]
y = df['label']
```

```python
# Train dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
#gausian aprroach
gnb = GaussianNB()
#multibinomial approach
mb =  MultinomialNB()
```

```python
#Trying to fit both gausian and multi binomial according to X and Y
gnb.fit(X_train,y_train)
mb.fit(X_train,y_train)

```

```python

y_expect = y_test
y_pred_gnb = gnb.predict(X_test)
y_predict_mb = mb.predict(X_test)
print(y_expect)
print(y_pred_gnb)
print(y_predict_mb)
```

```python
#accuracy rate
print(accuracy_score(y_expect,y_pred_gnb))
print(accuracy_score(y_expect,y_predict_mb))
```
