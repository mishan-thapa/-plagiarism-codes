```python
import random

def select_random_paragraphs(filename, min_length, max_length):
    with open(filename, "r",encoding = 'utf-8') as file:
        contents = file.read()
        paragraphs = re.split(r"[।|?|]",contents)
        selected_paragraphs1 = []
        for paragraph in paragraphs:
            if len(paragraph) >= min_length and len(paragraph) <= max_length :
                selected_paragraphs1.append(paragraph)
        if len(selected_paragraphs1) < 2:
            return []
        selected_paragraphs1 = random.sample(selected_paragraphs1, 50)
        return selected_paragraphs1

filename = "thesis/converted_1.txt"
selected_paragraphs1 = select_random_paragraphs(filename, min_length=15,max_length=200)
#print(selected_paragraphs1)
for paragraph in selected_paragraphs1:
    print(paragraph)
    print(len(paragraph))
```

```python
def preprocess_file(filename):
    with open(filename, "r",encoding = 'utf-8') as file:
        contents = file.readlines()
        processed_contents = []
        for line in contents:
            if line.strip().startswith(("१", "२","३","४","५","६","७","८","९")):
                continue
            processed_contents.append(line)
        return processed_contents

```

```python
filename = "thesis/converted_1.txt"
processed_contents = preprocess_file(filename)
```

```python
import random
```

```python
def select_random_sentences(filename,min_length, max_length):
    with open(filename, "r",encoding = 'utf-8') as file:
        contents = file.read()
        sentences = contents.split("।")
        selected_sentences = []
        for sentence in sentences:
            sentence = sentence.strip()
            if len(sentence) >= min_length and sentence[0]!="'" and sentence[0]!="‘" and sentence[0]!="‘‘"  and len(sentence) <= max_length:
                selected_sentences.append(sentence)
        if len(selected_sentences) < 2:
            return []
        selected_sentences = random.sample(selected_sentences, 100)
        return selected_sentences
```

```python
filename = "text/text1.txt"
selected_sentences1 = select_random_sentences(filename, min_length = 80, max_length = 150)
```

```python
filename = "text/text2.txt"
selected_sentences2 = select_random_sentences(filename, min_length = 80 , max_length = 150)
```
